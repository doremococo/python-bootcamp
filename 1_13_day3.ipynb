```python
# list に格納されているすべての数を平均する関数を定義。作成した関数を使い、他のリストの平均を算出。
import random
data_ls = [random.randint(1, 10000) for i in range(1, 11)]

def calculate_average(ls):
  return sum(ls) / len(ls)

print(f"平均値：{calculate_average(data_ls)}")


```
