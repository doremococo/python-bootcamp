```python
# input() はどんな時も文字列を返す。intでラップして数値に変換する。
inputNum = int(input("整数を入力してください:"))
if inputNum > 0:
  print("正の数です")
elif inputNum < 0:
  print("負の数です")
elif inputNum == 0:
  print("0です")
```
