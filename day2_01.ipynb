```python
ls01 = [1, 3, 5]
ls02 = [2, 4, 6]

# listの結合　＋:
ls_marge = ls01 + ls02
print("listを+で結合:", ls_marge)

# list結合　extend
ls03 = [11, 12, 13]
ls_marge.extend(ls03)
print("list結合　extend:", ls_marge)

```
