```python
#リスト[1,2,3]から要素をまとめて取り出し、変数a,b,cに格納したあと、変数a,cを出力

ls = [1, 2, 3]
dict_abc = {"a": 0, "b": 0, "c": 0}

# シンプルに実装
a = ls[0]
b = ls[1]
c = ls[2]
print(a, c)


# 繰り返しを使って実装
for i, (key, value) in enumerate(dict_abc.items()):
  dict_abc[key] = ls[i]

print(dict_abc["a"], dict_abc["c"])



```
