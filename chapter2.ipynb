```python
# as xxxx で、エイリアスを設定できる
import pandas as pd

# 2-01 pandasでcsv を読み込み、表示させる。
df = pd.read_csv("/content/adult_income_dataset.csv")
# df ←これだけで表示ができる
```

```python
# 2-02 dfの最初の5行目の表示と、行数、列数、データ型をそれぞれ確認

# 試しに head()を使ってみる　→ 最初の5行を取得できた
# df.head()

# 行数、列数、データ型の確認
# データ型の確認
df.dtypes

# 行と列の取得
# print(f"（行、列）：{df.shape}")
```

```python
# 2-03 df のage列を抽出し、先頭5行を出力
# df[列名]で特定の列だけを抽出する
df["age"].head()

# 複数の列を指定したい場合は、リスト内リスト的なやつで指定する
select_columns = ["income", "race"]
df[select_columns].head()

# df の列名を、リスト(要素は全て文字列※object) で取得
print(df.columns)
```

```python
# df [age] の値が30以上の行にfilterし、表示
# これをすると、ageの各行でboolean値が出る
df["age"] >= 30

# 試しにdfの中に先ほどのboolean値を入れてみたら出来た
df[df["age"] >= 30]

# もし「30歳以上」かつ「性別が女性（Female）」の人だけ出したいならこうする
df[(df["age"] >= 30) & (df["sex"] == "Female")]
# >> df[(df[列名]の条件式) & (df[列名]の条件式)] ← and = & , or = | です。それぞれの条件は（）で囲うこと。
```

```python
# 2-05 dfの[age] を昇順にソートして最初の5行を表示。by= は省略可能
df.sort_values(by="age").head()

# さらに fnlwgt の条件も追加する場合
df.sort_values(by=["age", "fnlwgt"]).head()
# >> 複数条件でソートしたい場合は、リストに条件を入れて by= に渡す。
```
