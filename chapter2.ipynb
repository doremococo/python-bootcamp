```python
# as xxxx で、エイリアスを設定できる
import pandas as pd

# 2-01 pandasでcsv を読み込み、表示させる。
df = pd.read_csv("/content/adult_income_dataset.csv")
# df ←これだけで表示ができる
df
```

```python
# 2-02 dfの最初の5行目の表示と、行数、列数、データ型をそれぞれ確認

# 試しに head()を使ってみる　→ 最初の5行を取得できた
# df.head()

# 行数、列数、データ型の確認
# データ型の確認
df.dtypes

# 行と列の取得
# print(f"（行、列）：{df.shape}")
```

```python
# 2-03 df のage列を抽出し、先頭5行を出力
# df[列名]で特定の列だけを抽出する
df["age"].head()

# 複数の列を指定したい場合は、リスト内リスト的なやつで指定する
select_columns = ["income", "race"]
df[select_columns].head()

# df の列名を、リスト(要素は全て文字列※object) で取得
print(df.columns)
```

```python
# df [age] の値が30以上の行にfilterし、表示
# これをすると、ageの各行でboolean値が出る
df["age"] >= 30

# 試しにdfの中に先ほどのboolean値を入れてみたら出来た
df[df["age"] >= 30]

# もし「30歳以上」かつ「性別が女性（Female）」の人だけ出したいならこうする
df[(df["age"] >= 30) & (df["sex"] == "Female")]
# >> df[(df[列名]の条件式) & (df[列名]の条件式)] ← and = & , or = | です。それぞれの条件は（）で囲うこと。
```

```python
# 2-05 dfの[age] を昇順にソートして最初の5行を表示。by= は省略可能
df.sort_values(by="age").head()

# さらに fnlwgt の条件も追加する場合
df.sort_values(by=["age", "fnlwgt"]).head()
# >> 複数条件でソートしたい場合は、リストに条件を入れて by= に渡す。
```

```python
# 2-06 欠損値の扱い方。変数dfに含まれる欠損値の数
print(df.shape)
# print(df["workclass"].isnull().sum())
# print(df[df["workclass"].isnull() == False].isnull().sum())

# dropna()で null になっている行を削除し、nullの数量を確認
df_cleaned = df.dropna()
df_cleaned.isnull().sum()
```

```python
# 2-07 特定列でグループ化する
age_group = df.groupby("age")

# groupは、dataflameとは別らしい。shapeが使えない
# NG → low_group = age_group.shape

# groupしたやつの列名を指定し、平均の時はmean()をする。複数列を計算したい時は、リスト内リストで渡す
ave_cg_fnl = age_group[["capital-gain", "fnlwgt"]].mean()
print(ave_cg_fnl)

# group の特定の列の集計において、複数の条件（合計、平均、最大など）をみたい時は agg(dict形式)で渡す
agg_cg_fnl= age_group.agg({"capital-gain": ["mean", "max"] ,"fnlwgt": ["sum", "median"]})
print(agg_cg_fnl)
```

```python
# 2-08 データを可視化する df に含まれる[age]列のヒストグラムを表示
df["age"].plot(
    kind="hist",
    title="age histgram",
    colormap="Accent",
    legend=True,
    grid=True
    )
```

```python
# 2-09 [age]列で平均値、中央値、標準偏差を求める
# 標準偏差は std()
df_mean = df["age"].mean()
df_median = df["age"].median()
df_std = df["age"].std()
print(df_mean, df_median, df_std)

# agg だとまとめて書ける
df_output = df["age"].agg(["sum", "mean", "std"])
print(df_output)
```
